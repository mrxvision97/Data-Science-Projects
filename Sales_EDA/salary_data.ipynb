```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("salary_data.csv")   # Reading CSV Files

df.shape    # Getting the Shape of the dataset
```

```python
df.head()   #printing the Head of the dataset
```

```python
df.info() #Gives the Information about the Attributes
```

```python
#checking for missing values
df.isnull().sum()
```

```python
df.describe()    #Extracting the summary of the dataset
```

```python
df1=df.groupby("continent_name")    #Grouping data based on column "Continent Name"
```

```python
df1.describe()   #Describing dataframe cotinent wise
```

From the above data, its clear that the North America, consisting of 4 countries has the highest average medain salary(4560.98) and Africa has the lowest average median salary (680.30). This signifies that North America is quite developed and Africa is under developed.

```python
#creating a plot
plt.figure(figsize=(15,6))
sns.scatterplot(data=df,x="continent_name",y="median_salary")
plt.show()
```

```python
#Drawing bar chart of median salary continent-wise
plt.figure(figsize=(15,6))
sns.barplot(data=df,x="continent_name",y="median_salary")
plt.title("Bar Chart of Median Salary")
plt.xlabel("Continents")
plt.ylabel("Median Salary")
plt.show()
```

```python
#drawing line plot to see the trend in global salary
plt.figure(figsize=(15,6))
sns.lineplot(data=df,x="continent_name",y="median_salary")
plt.title("Median Salary of each continent")
plt.show()
```

```python
highest_salary=df[df["median_salary"]==df["median_salary"].max()]
print("Country with highest median_salary :\n",highest_salary)
```

```python
lowest_salary=df[df["median_salary"]==df["median_salary"].min()]
print("Country with highest median_salary :\n",lowest_salary)
```
